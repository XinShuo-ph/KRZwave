```python
from KRZmetric import *
import numpy as np
import matplotlib.pyplot as plt
#观察Kerr下调Ma有没有isofrequency pairing
```

```python
e=0.5
p=6
```

```python
Mlist=np.linspace(1e6,4e6,21)
alist=np.linspace(0.6,1.0,21)
```

```python
print(Mlist)
print(alist)
```

```python
Mgrid=np.zeros((len(Mlist),len(alist)))
agrid=np.zeros((len(Mlist),len(alist)))
omgrgrid=np.zeros((len(Mlist),len(alist)))
omgphigrid=np.zeros((len(Mlist),len(alist)))
for i in np.arange(len(Mlist)):
    for j in np.arange(len(alist)):
        Mgrid[i,j]=Mlist[i]
        agrid[i,j]=alist[j]
        omgrgrid[i,j],omgphigrid[i,j]=getfreq_sec_fromepma(e,p,Mgrid[i,j],agrid[i,j])
        print("%d,%d, M=%f, a=%f, omgr=%f, omgphi=%f"%(i,j,Mgrid[i,j],agrid[i,j],omgrgrid[i,j],omgphigrid[i,j]))
```

```python
myma=0.01
plt.contour(omgphigrid,agrid,omgrgrid,cmap='jet')
plt.xlim(0.0007,myma)
plt.ylim(0.6,1.0)
plt.text(myma*0.7,0.9,'e=%.1f, p=%.1f'%(e,p),fontsize=16)
plt.xlabel('$\Omega_\phi$',fontsize=20)
plt.ylabel('a',fontsize=20,rotation=0)
plt.savefig('test_a-isofreq.png')
```
