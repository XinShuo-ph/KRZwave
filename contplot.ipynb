```python
fig1,ax=plt.subplots()
CS=ax.contourf(pg,eg,FF,cmap='jet')
cbar = fig1.colorbar(CS)
cbar.ax.set_ylabel('overlap')
ax.arrow(6.25,0.425,0.15,-0.015,head_width=0.01,shape='right',color='white')
ax.text(6,0.43,'same $\omega$',color='white')
ax.set_xlabel('$p_{Kerr}$/M')
ax.set_ylabel('$e_{Kerr}$')

fig1.savefig('FF1.pdf')
```
